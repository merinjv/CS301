Exercise 3a

```python
import numpy as np

count_outcome_two = 0;
number_of_trials = 1000;

for n in range(number_of_trials):
  outcome = np.random.randint(1,11); #random number between 1-10
  count_outcome_two+=np.sum((outcome==2)*1); 

p = count_outcome_two/(number_of_trials);
print("Probability of outcome 2: ",p);
```

Exercise 3b

```python
import matplotlib.pyplot as plt

prob = [];
trials = [];
plt.xlabel('Max Number of Trials (n)');
plt.ylabel('Probability (p)');
for n in range(10, number_of_trials+10,10):
  outcome = np.random.multinomial(n,[1/10.]*10,1);
  p = (outcome[0][1])/(number_of_trials);
  prob.append(p);
  trials.append(n);
  plt.plot(trials,prob);

print('As n increases the probability (p) also increases.')
```
