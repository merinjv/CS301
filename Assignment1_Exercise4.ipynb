```python
import matplotlib.pyplot as plt
import numpy as np

figure = plt.figure(figsize=(10,10));
figure.tight_layout();

#plot1

mean = [0,0];
cov = [[1,0], [0,1]];
x,y = np.random.multivariate_normal(mean, cov, 1000).T;
plt.subplot(2,2,1);
plt.plot(x,y,'x');
plt.title('p = 0.0');

#plot2

mean = [0,0];
cov = [[1,0.2], [0.2,1]];
x,y = np.random.multivariate_normal(mean, cov, 1000).T;
plt.subplot(2,2,2);
plt.plot(x,y,'x');
plt.title('p = 0.20');

#plot3

mean = [0,0];
cov = [[1,-0.8], [-0.8,1]];
x,y = np.random.multivariate_normal(mean, cov, 1000).T;
plt.subplot(2,2,3);
plt.plot(x,y,'x');
plt.title('p = -0.80');

#plot4

mean = [0,0];
cov = [[1,0.99], [0.99,1]];
x,y = np.random.multivariate_normal(mean, cov, 1000).T;
plt.subplot(2,2,4);
plt.plot(x,y,'x');
plt.title('p = 0.99');
```
